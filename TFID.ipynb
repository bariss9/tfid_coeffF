## TF-ID ve TF-ID Vectorize nedir?

Tfıd -> bir kelimenin bir belge içerisindeki önemini ölçer.Belgede sık geçen ama genel olarak anlamsız olan kelimeleri arka plana atıp önemli olanları öne çıkarır.

Tfıd Vectorize -> Metinleri sayısallaştırarak ml'e uygun hale getirir, sık geçen ama anlamsız olan kelimeleri azaltır

Vectorize -> belgedeki kelimeleri sayısallaştırır ve cümlelerde olanları sayısal olarak gösterir

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
import pandas as pd

documents = [
    "kedi maması çok lezzetli",
    "köpek maması çok lezzetli",
    "lezzetli yemek seviyorum"
]
```

```python
count_vectorizer = CountVectorizer() # kelime sıklığını ölçen fonksiyon
count_matrix = count_vectorizer.fit_transform(documents)

count_df = pd.DataFrame(count_matrix.toarray(), columns=count_vectorizer.get_feature_names_out())
print(" CountVectorizer Çıktısı:") # düz vectorize sonuçları tablo haline getirdik
display(count_df)

tfidf_vectorizer = TfidfVectorizer() #burda tfıd vectorizer önemli olan kelimeleri ön plana çıkarır
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
print(" TF-IDF Vectorizer Çıktısı:") # tfıd vectorize sonuçları tablo halinde
display(tfidf_df)
  
```
